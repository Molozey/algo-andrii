```python
from sklearn import model_selection, ensemble, metrics, datasets, tree
import numpy as np
import matplotlib.pylab as plt
```

```python
raw_data = datasets.load_digits()
```

```python
data = raw_data.data
targets = raw_data.target
```

```python
decision_tree_class = tree.DecisionTreeClassifier()
cross_score = model_selection.cross_val_score(decision_tree_class, data, targets,
                                              scoring='accuracy', cv=10, n_jobs=2)
```

```python
cross_score.mean(axis=0)
def write_answer_1(ans, number):
    with open("answer"+str(number)+".txt", "w") as text_file:
        text_file.write(str(ans))
write_answer_1(cross_score.mean(axis=0),1)
```

```python
bagging = ensemble.BaggingClassifier(base_estimator=decision_tree_class, n_estimators=100)
cross_score = model_selection.cross_val_score(bagging, data, targets, scoring='accuracy',
                                              cv=10, n_jobs=1)
```

```python
print(cross_score.mean(axis=0))
write_answer_1(cross_score.mean(axis=0),2)
```
