# Кросс-валидация

```python
from sklearn import datasets
from sklearn import model_selection
import numpy as np
```

```python
iris = datasets.load_iris()
train_data, test_data, train_labels, test_labels = model_selection.train_test_split(
    iris.data, iris.target, test_size=0.3
)
```

```python
print('Размер обучающей выборки {} объектов \n Размер тестовой выборки {} объектов'.format(len(train_data), len(test_data)))
```

## Стратегии проведения кросс-валидации

### KFold
Каждая выборка разбивается на k блоков каждый блок k-1 раз участвует в обучении и 1 раз в тестировании
##### Shuffle=True перемешивает выборку
#### Random_state = int Фиксирует случайность перемешивания выборки

```python
cross_valid = model_selection.KFold(2, shuffle=True, random_state=1)
cross_valid.get_n_splits(iris.data, iris.target)
for train_cross, test_cross in cross_valid.split(iris.data, iris.target):
    print('TRAIN:', iris.data[train_cross], 'TRAIN TARGET', iris.target[test_cross])
    print('TEST:', iris.data[test_cross], 'TEST TARGET:', iris.target[test_cross])
```

### StratifiedKFold
Сохраняется соотношение классов в обучающей и тестовой выборке

### ShuffleSplit
Строит случайные перестановки
n_iter = количество разных блоков

```python
shuffle_valid = model_selection.ShuffleSplit(n_splits=50, test_size=0.2)
shuffle_valid.get_n_splits(iris.data, iris.target)
for train_cross, test_cross in shuffle_valid.split(iris.data, iris.target):
    print('TRAIN DATA:')
    print(iris.data[test_cross])
    print('TRAIN TARGET:')
    print(iris.target[test_cross])
    print('TEST DATA:')
    print(iris.data[test_cross])
    print('TEST TARGET:')
    print(iris.target[test_cross])
```

### LeaveOneOut

Каждый объект побудет в выборке по одному разу
