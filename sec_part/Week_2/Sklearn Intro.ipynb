### sklearn.circles

```python
circles = datasets.make_circles()
```

```python
print ("feathures: {}".format(circles[0][:10]))
print ("target: {}".format(circles[1][:10]))
```

```python
from matplotlib.colors import ListedColormap
from matplotlib import pyplot
```

```python
colors = ListedColormap(['red', 'yellow'])
pyplot.figure(figsize=(8,8))
pyplot.scatter(list(map(lambda x: x[0], circles[0])), list(map(lambda x: x[1], circles[0])), c=circles[1], cmap=colors)
```

```python
def data_cir_2d(circles, colors):
    pyplot.figure(figsize=(8,8))
    pyplot.scatter(list(map(lambda x: x[0], circles[0])), list(map(lambda x: x[1], circles[0])), c=circles[1], cmap=colors)
```

```python
noise_circles = datasets.make_circles(noise=0.15)
data_cir_2d(noise_circles, colors)
```

### sklearn.classification

```python
simple_classification = datasets.make_classification(n_features=2, n_informative=1, n_redundant=1,
                                                     n_clusters_per_class=1, random_state=1)
```

```python
data_cir_2d(simple_classification, colors)
```

```python
classification = datasets.make_classification(n_features=2, n_informative=2, n_classes=4,
                                              n_redundant=0, n_clusters_per_class=1, random_state=1)
colors = ListedColormap(['red', 'blue', 'green', 'yellow'])
```

```python
data_cir_2d(classification, colors)
```

```python
iris = datasets.load_iris()
iris
```

```python
iris.keys()
print(iris.DESCR)
```

```python
from pandas import DataFrame
iris_pd = DataFrame(iris.data)
iris_pd.columns = iris.feature_names
iris_pd['target'] = iris.target
iris_pd.target= iris_pd.target.apply(lambda x: iris.target_names[x])
```

```python
iris_pd
```

```python
iris_pd[iris_pd.target == 'setosa'].hist('sepal length (cm)')
```

```python
pyplot.figure(figsize=(12,12))
plot_arrow = 0
for feature_name in iris['feature_names']:
    for target_name in iris['target_names']:
        plot_arrow += 1
        pyplot.subplot(4,3, plot_arrow)
        pyplot.hist(iris_pd[iris_pd.target == target_name][feature_name])
        pyplot.title(target_name)
        pyplot.xlabel('cm')
        pyplot.ylabel(feature_name[:-4])
```
