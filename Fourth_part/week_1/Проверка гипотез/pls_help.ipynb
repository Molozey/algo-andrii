```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 16
F_H0 = stats.binom(n, 0.5)
```

```python
x = np.linspace(0, 16, 17)
plt.bar(x, F_H0.pmf(x), align='center')
plt.xlim(-.5, 16.5)
plt.show()
```

```python
F_H0.pmf(x)
```

```python
#   Одностороняя альтернатива
plt.bar(x, F_H0.pmf(x), align='center')
plt.bar(np.linspace(12,16,5), F_H0.pmf(np.linspace(12,16,5)), align='center', color='red')
plt.xlim(-.5, 16.5)
plt.show()
stats.binom_test(12, 16, 0.5, 'greater')
```

```python
#   Одностороняя альтернатива
plt.bar(x, F_H0.pmf(x), align='center')
plt.bar(np.linspace(11,16,6), F_H0.pmf(np.linspace(11,16,6)), align='center', color='red')
plt.xlim(-.5, 16.5)
plt.show()
stats.binom_test(11, 16, 0.5, 'greater')
```

```python
#   Двустороняя альтернатива
plt.bar(x, F_H0.pmf(x), align='center')
plt.bar(np.linspace(0,4,5), F_H0.pmf(np.linspace(0,4,5)), align='center', color='red')
plt.bar(np.linspace(12,16,5), F_H0.pmf(np.linspace(12,16,5)), align='center', color='red')
plt.xlim(-.5, 16.5)
plt.show()
stats.binom_test(12,16,.5, alternative='two-sided')
```
