```python
import numpy as np
import matplotlib.pyplot as plt

a=np.linspace(-5,5,100)
b=np.linspace(-5,5,100)
A,B = np.meshgrid(a,b)
z=((0.2 * A + 1)**2 + (0.2 * B) ** 2)/((1- 0.8 * A) ** 2+ (0.8*B) ** 2)
plt.figure(figsize=(6, 6))
plt.contour(a,b,z,[1])
ax = plt.gca()
ax.tick_params(labelsize=18)
plt.grid()
plt.show()
```
