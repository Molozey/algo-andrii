```python
import matplotlib.pylab as plt
import numpy as np
import scipy.stats as sts
from collections import Counter
```

## Пуассона

```python

normal_stat = sts.poisson(3)
normal_stat.rvs(5)
x = np.linspace(0,30,100)
plt.title('Частота распределения')
plt.hist(normal_stat.rvs(1000), bins=10, color='orange', density=True, histtype='bar')
plt.plot(x, normal_stat.pmf(x), linewidth=3)
#plt.hist(normal_stat.pdf(x), bins=20, density=True)
plt.show()

def colorize(n):
    if (n % 10) == 1:
        return 'orange'
    if (n % 10) == 2:
        return 'black'
    if (n % 10) == 3:
        return 'red'
    if (n % 10) == 4:
        return 'pink'
    if (n % 10) == 5:
        return 'blue'
    if (n % 10) == 6:
        return 'red'
    if (n % 10) == 7:
        return 'yellow'
    if (n % 10) == 8:
        return 'blue'
    if (n % 10) == 9:
        return 'pink'
    if (n % 10) == 0:
        return 'red'
```

### Делаем 1000 выборок

```python
def viborka_1000(n):
    normal_stat = sts.poisson(3)
    numbers = []
    average_array = []
    for counter in range(1, 1001):
        random_array = normal_stat.rvs(n)
        average = np.average(random_array)
        numbers.append(counter)
        average_array.append(average)
    plt.hist(average_array, bins=40, label='Histogram for n='+str(n), color=colorize(n), density=True)
    plt.legend(loc='upper right')

def hist_vib(n):
    k = n
    viborka_1000(n)
    nu = 3
    sigma = 3 / k
    normal_stat = sts.norm(loc=nu, scale=np.sqrt(sigma))
    x = np.linspace(-4,4,100)
    plt.plot(x, normal_stat.pdf(x), linewidth=3)
```

```python
hist_vib(5)
```
