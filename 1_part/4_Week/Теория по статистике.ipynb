### В этом блокноте содержатся первоначальные сведения о статистиках


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from statsmodels.distributions.empirical_distribution import ECDF
```

### Дискретное распределение. Броски кубика

```python
sample = np.random.choice([1,2,3,4,5,6], 100)
from collections import Counter
c = Counter(sample)
c = sorted(c.items())
for i in range(len(c)):
    plt.plot(list(c[i])[0], list(c[i])[1], 'o')
```

```python
c = Counter(sample)
print({k: v/100 for k,v in c.items()})
```

### Непрерывное распределение

```python
normal = sts.norm(0,1)
sample = normal.rvs(100)

array = np.arange(-4,4,0.1)
plt.plot(array, normal.cdf(array), label='THEOR CDF')
ecdf = ECDF(sample)
plt.plot(ecdf.x,ecdf.y)
plt.legend(loc='upper left')
```

```python
plt.hist(sample, bins=30, density=True)
```

### Ядерное сглаживание

```python
import pandas as pd
df = pd.DataFrame(sample, columns=['KDE'])
df.plot(kind='density')
array = np.arange(-4,4,0.1)
plt.plot(array, normal.pdf(array), label='THEOR CDF')
plt.legend(loc='upper left')
```

```python
p = 0.38
n = 100
low_bord = p - 2 * np.sqrt(p * (1 - p) / n)
hight_bord = p + 2 * np.sqrt(p * (1 - p) / n)
print(low_bord)
print('---')
print(hight_bord)
```
