```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
```

## Нормальное распределение из Скиппи
.pdf - показывает значение распределения плотности вероятности

.cdf показывает значение интеграла посчитанного от 0 до указанного параметра

.rvs - значние для n параметров

```python
nu = 2
sigma = 0.5
normal_func = sts.norm(loc=nu, scale=sigma)
x = np.linspace(0,5,100)
plt.plot(x, normal_func.cdf(x))
```

## Равномерное распределение
scipy.stats.uniform важно что первый аргумент - начало, второй это расстояние до начала

Фактически второй = КОНЕЦ - НАЧАЛО

```python
x0 = 2
x1 = 10
equal_func = sts.uniform(x0 , x1-x0)
plt.plot(np.linspace(x0 - 4,x1 + 4,100), equal_func.pdf(np.linspace(x0 - 4,x1 + 4,100)))
```

## Распределение Бернулли
scipy.stat.bernoulli(p)
p - вероятность успеха

```python
bernuli = sts.bernoulli(.3)
plt.grid()
plt.plot(bernuli.rvs(10),'o')
```

## Биномиальное распределение
sts.binom(n,p)
n - количество экспериментов
p - вероятность успеха
Сам биноминал показывает количество успехов из выборки n эксперементов с вероятностью p
Скорей дает распределение вероястность получить именно k успехов из n проведенных экспериментов


Суть в том что чем больше берем точек, то тем больше будет успехов
cdf - показывает вероятность того количество успехов будет n и больше
.pmf - Аналог для функции плотности .pdf. Показывает какое количество успехов будет иметь кааую вероястность

```python
binomial = sts.binom(20, .7)
array = np.linspace(1, 20, 20)
#plt.step(array, binomial.cdf(array))
plt.plot(array, binomial.pmf(array),'o')
```

### Распределение Пуассона
sts.poisson(lamda)

```python
poisson = sts.poisson(20)
array = np.linspace(0,30, 31)
plt.plot(array, poisson.cdf(array))
```
