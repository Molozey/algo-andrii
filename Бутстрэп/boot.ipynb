```python
import pandas as pd
import numpy as np


import yfinance as finance

from scipy import stats
import matplotlib.pylab as plt
import seaborn as sns
```

```python
sber = finance.download('SBER.ME', start='2006-01-01', end="2021-01-01")
with sns.axes_style('white'):
    sber.plot(subplots=True, figsize=(14, 8), title=['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], legend=False)
```

```python
sber_pre = sber["Close"].pct_change().dropna()
mu = sber_pre.mean()
sigma = sber_pre.std()
normed = stats.norm(mu, sigma)
plt.figure(figsize=(10,8))
plt.hist(normed.rvs(2275), bins=1000, color='blue')
plt.hist(sber_pre, bins=1000, color='red', alpha=.3)
plt.xlim(-.25, .25)
plt.show()
```
